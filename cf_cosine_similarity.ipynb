```python
import pandas as pd
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity
```

```python
initial = pd.read_csv('static/data/ratings_pivot_table.csv', index_col=0)
initial = initial.T # transpose
initial
```

```python
user_item = initial.fillna(0)
user_item
```

```python
user_similarity = cosine_similarity(user_item.T)
user_similarity = pd.DataFrame(user_similarity, columns = user_item.columns, index = user_item.columns).round(2)
user_similarity
```

```python
movies = initial.index
movies
```

```python
new_user_query = {'Fight Club (1999)': 5}
df_new_user = pd.DataFrame(
        data=new_user_query,
        columns=movies,
        index = ['new_user']
)
df_new_user.T
```

```python
user_item_with_new_user = user_item.join(df_new_user.T).fillna(0)
user_item_with_new_user
```

```python
user_similarity = cosine_similarity(user_item_with_new_user.T)
user_similarity = pd.DataFrame(user_similarity, columns = user_item_with_new_user.columns, index = user_item_with_new_user.columns).round(2)
user_similarity
```

```python
candidates = movies[df_new_user.T['new_user'].isna()]
candidates
```

```python
neighbours = user_similarity['new_user'].sort_values(ascending=False).index[1:6]
neighbours
```

```python
for movie in candidates:
    # extract users who have rated unseen movies
    current_movie_reviewers = initial.columns[~initial.loc[movie].isna()]
    # we want to create an intersection, so we save the users as a set-object
    current_movie_reviewers = set(current_movie_reviewers)
```

```python
predictions = {}

for movie in candidates:
    #extract users who have rated unseen movies
    current_movie_reviewers = initial.columns[~initial.loc[movie].isna()]
    # we want to create an intersection, so we save the users as a set-object
    current_movie_reviewers = set(current_movie_reviewers)

    intersection = set(neighbours).intersection(current_movie_reviewers)
    if len(intersection) != 0:
        num = 0
        den = 0
        for neighbour in intersection:
            ratings = user_item_with_new_user[neighbour][movie] 
            sim_score = user_similarity['new_user'][neighbour]
            num = num + (ratings*sim_score)
            den = den + sim_score
        
        pred_ratings = num/den
        predictions[movie] = pred_ratings

predictions
```

```python
[movie for movie, _ in sorted(predictions.items(), key=lambda item: item[1], reverse=True)][:10]
```
